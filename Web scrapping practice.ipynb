```python
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import logging


## needed libraries above

```

```python
### url build up for query 
flipcart_url = "https://www.flipkart.com/search?q=" + "tv"
```

```python
flipcart_url
```

```python
## open url remotely

urlclient = urlopen(flipcart_url)
```

```python
urlclient
```

```python
flipcart_page = urlclient.read()
```

```python
flipcart_page  ### content of the page
```

```python
## Beautify the page make more human readable
flip_html = bs(flipcart_page, "html.parser") ## using module beautifulsoup
```

```python
flip_html
```

```python
## use find_all to aggregate the content which has same class names

bigbox = flip_html.find_all("div", {"class":"_1AtVbE col-12-12"})
```
